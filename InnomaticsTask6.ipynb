```python
# Program 1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    current = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1
        
        if current > peak:
            peak = current
    
    status = "Full" if current >= capacity else "Available"
    
    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
# Program 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5
    
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8
    
    if weather == "Rainy":
        base_time += 10
    elif weather == "Stormy":
        base_time += 20
    
    print("Estimated Delivery Time:", base_time, "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

```python
# Program 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"
    
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)

total_seats = 200
booked = [1] * 150

analyze_occupancy(total_seats, booked)
```

```python
# Program 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0
    
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)

cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

```python
# Program 5: Smart Classroom Resource Usage Monitor

def monitor_resources(resource_usage):
    overused = []
    
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    alert = "Yes" if overused else "No"
    
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resources)
```

```python
# Program 6: Online Event Registration Capacity Controller

def manage_registrations(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

manage_registrations(100, 105)
```

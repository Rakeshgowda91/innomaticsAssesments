```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return f"Invalid plan selected. Available plans: {valid_plans}"
    else:
        return "Recharge successful!"

recharge_amount = 299
print(validate_recharge(recharge_amount))
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} → Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} → Stock OK (Stock: {stock})")

inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 8,
    "Monitor": 25
}

check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"

student_marks = [60, 45, 70, 50]
print(process_result(student_marks))
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak:
        fare += fare * 0.25
    
    return fare

distance = 10
peak_hour = True

print(f"Total Fare: ₹{calculate_fare(distance, peak_hour):.2f}")
```

```python
def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
def check_password(password):
    special_characters = "@#$"
    
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return "Weak Password"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = "Data@123"
print(check_password(password))
```

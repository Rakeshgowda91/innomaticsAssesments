```python
# Program 1: Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like
    
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

```python
# Program 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock(stock):
    status = "Low Stock Alert" if stock < 10 else "Stock Sufficient"
    
    print("Medicine Stock:", stock)
    print("Status:", status)

check_medicine_stock(6)
```

```python
# Program 3: Agriculture – Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"
    
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall = [70, 75, 80, 63]
check_rainfall(rainfall, 70)
```

```python
# Program 4: Social Media – Duplicate Account Detection

def detect_duplicates(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"
    
    print("Duplicate Accounts Found:", result)

users = ["john", "emma", "john", "alex"]
detect_duplicates(users)
```

```python
# Program 5: Healthcare – Appointment Eligibility Checker

def check_appointment_eligibility(age):
    status = "Eligible" if age >= 18 else "Not Eligible"
    
    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_appointment_eligibility(21)
```

```python
# Program 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Program 7: System Monitoring – Application Health Checker

def check_system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

check_system_health(7)
```

```python
# Program 8: Banking – Daily Transaction Limit Checker

def check_transaction(amount):
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

check_transaction(60000)
```

```python
# Program 9: E-Learning – Student Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    present = 0
    
    for day in attendance_list:
        if day == "P":
            present += 1
    
    percentage = (present / len(attendance_list)) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

```python
# Program 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    print("Total Bill:", bill)
    print("Usage Level:", usage)

calculate_electricity_bill(250)
```

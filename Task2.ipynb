```python
s = "python is easy and python is powerful"

words = s.split()
u = set(words)

print("Unique words count:", len(u))
print("Unique words:", u)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

name = ""
high = 0

for k in employees:
    if employees[k] > high:
        high = employees[k]
        name = k

print("Highest Salary:", name, "-", high)
```

```python
numbers = [45, 22, 89, 10, 66]

mx = numbers[0]
mn = numbers[0]

for n in numbers:
    if n > mx:
        mx = n
    if n < mn:
        mn = n

print("List:", numbers)
print("Max:", mx)
print("Min:", mn)
```

```python
prices = [450, 1200, 899, 1500, 300]

c = 0

for p in prices:
    if p > 1000:
        c = c + 1

print("Products above 1000:", c)
```

```python
attendance = ["P", "P", "A", "P", "P"]

p = 0

for a in attendance:
    if a == "P":
        p = p + 1

percentage = (p / len(attendance)) * 100

print("Attendance Percentage:", percentage)
```

```python
s = "python"

d = {}

for ch in s:
    if ch in d:
        d[ch] = d[ch] + 1
    else:
        d[ch] = 1

print(d)
```

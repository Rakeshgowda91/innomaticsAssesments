```python
marks = [45, 78, 90, 33, 60]

p = 0
f = 0

for m in marks:
    if m >= 50:
        p = p + 1
    else:
        f = f + 1

print("Pass:", p)
print("Fail:", f)
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

c = 0

for i in logs:
    if i == "ERROR":
        c = c + 1

print("Errors:", c)
```
